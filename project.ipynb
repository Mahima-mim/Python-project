```python
#creating a pandas dataFrame
import pandas as pd
df= pd.read_csv('environmental_data.csv')
print(df)
```

```python
#Barplot using Seaborn

import seaborn as sns
import matplotlib.pyplot as plt

#sat seaborn style
sns.set_style('darkgrid')

#Set Seaborn color palatte
sns.set_palette('deep')

#Set Seaborn plot

sns.barplot(x='Temperature',y='Humidity',data=df)


#Adding labels

plt.xlabel('Temperature',fontsize=14,fontweight='bold')

plt.ylabel('Humidity',fontsize=14,fontweight='bold')

plt.title('Bar plot using pandas DataFrame',fontsize=16,fontweight='bold')

plt.show()
```

```python
#Data visualisation using bokeh
#Line graph

from bokeh.plotting import figure,show,output_notebook
from bokeh.models import HoverTool

output_notebook()

p=figure(title="Simple Bokeh Plot",x_axis_label="Temperature",y_axis_label='Pressure')


#figure built in function
p.line(x='Temperature', y='Humidity', source=df, line_width=5, color='RebeccaPurple', legend_label="Humidity Trend")
p.scatter(x='Temperature', y='Humidity', source=df, size=8, fill_color="Gold", line_width=1, legend_label="Data Points")

#Add hover tool

hover=HoverTool()
hover.tooltips=[('Temperature','$Temperature'),('Pressure','$Pressure')]
p.add_tools(hover)

show(p)
```
